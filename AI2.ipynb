```python
import pandas as pd
url = 'medals.csv'
dataframe = pd.read_csv(url)
print(dataframe)
```

Chọn quốc gia để báo cáo bài tập là nước Pháp (FRA)

Số huy chương vàng của Pháp (FRA) qua các năm :


```python
dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Gold')].head(1000)
```

```python
fram = dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Gold')].count()
print("Số huy chương vàng đạt được của Pháp qua các năm:",fram['Medal'])
```

Số huy chương bạc của Pháp(FRA) qua các năm :

```python
dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Silver')].head(1000)
```

```python
fram = dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Silver')].count()
print("Số huy chương bạc đạt được của Pháp qua các năm:",fram['Medal'])
```

Số huy chương đồng của Pháp (FRA) qua các năm :

```python
dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Bronze')].head(1000)
```

```python
frame = dataframe[(dataframe['NOC'] == 'FRA')&(dataframe['Medal'] == 'Bronze')].count()
print("Số huy chương đồng đạt được của Pháp qua các năm:",frame['Medal'])
```

Tổng số huy chương theo từng bộ môn của Pháp qua các năm:

```python
Sport = dataframe[dataframe['NOC']=='FRA']
Sport.groupby('Sport')['Medal'].count()
```

```python
Sport = dataframe.loc[dataframe.NOC == 'GER']
total = Sport.Sport.value_counts().count()
print('Tổng số bộ môn mà GER tham gia qua tất cả các năm : ',total)
```

Các năm Pháp tham dự Olympics mùa đông và có huy chương:

```python
frame = dataframe.loc[dataframe.NOC == 'FRA']
numberofyear = frame['Year'].unique()
print('Các năm tham dự Olympics mùa đông của Pháp : ',numberofyear)
total = frame.Year.value_counts().count()
print('\n Tổng số năm tham dự Olympics của Pháp : ',total)
```
